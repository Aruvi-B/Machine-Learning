```python
df.replace("?", np.nan, inplace = True)
# df.dropna(axis = 0, inplace = True)
```

```python

for column in df.columns:
  if df[column].dtype == np.float64 or df[column].dtype == np.int64:
    df[column].fillna(df[column].mean(), inplace=True)
  elif df[column].dtype == np.object:
    df[column].fillna(df[column].mode()[0], inplace=True)
```

```python
from sklearn import preprocessing
label = preprocessing.LabelEncoder()
df["make"] = label.fit_transform(df["make"])
df["fuel-type"] = label.fit_transform(df["fuel-type"])
df["engine-location"] = label.fit_transform(df["engine-location"])
df["aspiration"] = label.fit_transform(df["aspiration"])
df["num-of-doors"] = label.fit_transform(df["num-of-doors"])
df["fuel-system"] = label.fit_transform(df["fuel-system"])
df["body-style"] = label.fit_transform(df["body-style"])
df["drive-wheels"] = label.fit_transform(df["drive-wheels"])
df["engine-type"] = label.fit_transform(df["engine-type"])
df["num-of-cylinders"] = label.fit_transform(df["num-of-cylinders"])
```

```python
df['normalized-losses'] = df['normalized-losses'].astype(float)
df['bore'] = df['bore'].astype(float)
df['stroke'] = df['stroke'].astype(float)
df['horsepower'] = df['horsepower'].astype(float)
df['peak-rpm'] = df['peak-rpm'].astype(float)
df['price'] = df['price'].astype(float)
df['make'] = df['make'].astype(float)
```

```python
for col in df.columns:
    if df[col].dtype == 'object':
        df.col = pd.to_numeric(df[col])
        df.col = df[col].astype(float)
        print(f"Non-numerical column: {col}")
```

```python
df.columns
```

```python
X = df[["make", "drive-wheels"]]
y = df['price']

from sklearn.preprocessing import LabelEncoder
labele = LabelEncoder()
y = labele.fit_transform(y)
```

```python
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, random_state = 42)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import mean_absolute_error
```

```python
from sklearn.tree import DecisionTreeRegressor
tree_reg = DecisionTreeRegressor()
tree_reg.fit(X_train, y_train)
y_pred_tree = tree_reg.predict(X_test)
mae_tree = mean_absolute_error(y_test, y_pred_tree)
print("Mean Absolute Error:", mae_tree)

```
