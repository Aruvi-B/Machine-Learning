```python
# prompt: this is ok, i need additional also like
# alter the CV column is (12.26 * CO) + (10.79 * H2) + (35.81 * CH4)
# apply the above formula for all the CV values

import numpy as np
import pandas as pd

# Define the components and their respective upper and lower limits
components = {
    "C": [44.181, 43.201],
    "H": [6.125, 6.102],
    "N": [0.55, 0.68],
    "S": [0, 0],
    "O": [53.156, 50.257],
    "MC": [4.487, 4.087],
    "ASH": [0.23, 0.204],
    "FIXED CARBON": [6.181, 6.158],
    "VM": [91.014, 89.418],
    "CGE": [71.158, 68.243],
    "CV": [6.842, 6.243],
    "CO": [23.454, 19.243],
    "CO2": [14.254, 13.451],
    "CH4": [3.542, 3.120],
    "H2": [20.156, 18.654],
}

# Generate 118 sets of values for each component within the specified range
np.random.seed(42)  # For reproducibility
data = {comp: np.random.uniform(low=limits[1], high=limits[0], size=118) for comp, limits in components.items()}

# Create a DataFrame from the generated data
df = pd.DataFrame(data)

# Display the first few rows of the generated data
df

def adjust_composition(row):
  """
  Adjusts the values of C, H, N, S, and O in a row to ensure their sum is between 99 and 99.989.
  """
  total = row["C"] + row["H"] + row["N"] + row["S"] + row["O"]

  # Calculate the adjustment factor
  if total > 99.989:
    factor = 99.989 / total
  elif total < 99:
    factor = 99 / total
  else:
    return row  # No adjustment needed

  # Apply the adjustment factor to the relevant columns
  row["C"] = row["C"] * factor
  row["H"] = row["H"] * factor
  row["N"] = row["N"] * factor
  row["S"] = row["S"] * factor
  row["O"] = row["O"] * factor
  return row

df = df.apply(adjust_composition, axis=1)

df["CV"] = ((12.26 * df["CO"]) + (10.79 * df["H2"]) + (35.81 * df["CH4"]))/100

df
```

```python
# prompt: some of the sum is 99.05, 99.63 like that i need but this sum is all 99.95
# so alter that values of column MC, ASH, FIXED CARBON, VM in minute manner
# if i add the values measn i get 99 to 99.95 range values

def adjust_secondary_components(row):
  """
  Adjusts the values of MC, ASH, FIXED CARBON, and VM in a minor way to ensure the sum of all components
  falls within the desired range (99 to 99.989).
  """
  primary_sum = row["C"] + row["H"] + row["N"] + row["S"] + row["O"]
  secondary_sum = row["MC"] + row["ASH"] + row["FIXED CARBON"] + row["VM"]
  total_sum = primary_sum + secondary_sum

  # Calculate the adjustment factor
  if total_sum > 99.989:
    factor = 99.989 / total_sum
  elif total_sum < 99:
    factor = 99 / total_sum
  else:
    return row  # No adjustment needed

  # Adjust secondary components
  row["MC"] = row["MC"] * factor
  row["ASH"] = row["ASH"] * factor
  row["FIXED CARBON"] = row["FIXED CARBON"] * factor
  row["VM"] = row["VM"] * factor
  return row

df = df.apply(adjust_secondary_components, axis=1)
```
